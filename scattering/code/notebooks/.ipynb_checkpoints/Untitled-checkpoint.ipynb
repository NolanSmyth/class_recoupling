```python
import sys
import matplotlib.pyplot as plt
import numpy as np

sys.path.append("./build/")
```

```python
from ts_rates import Model, thermal_scattering_rate
```

```python
m = Model(1.0, 25.0, 1.0, 1.0)
```

```python
thermal_scattering_rate(1.0, m)
```

```python
params = {"dm": 1.01, "r": 25.0, "g": 1e-2, "lam": 1e4}
m1 = Model(**params)

params = {"dm": 1.0, "r": 25.0, "g": 1e-3, "lam": 1e6}
m2 = Model(**params)

Ts = np.geomspace(1e-3, 1e3, 100)

plt.figure(dpi=100)
plt.plot(Ts, np.array([thermal_scattering_rate(T, m1) for T in Ts]))
plt.plot(Ts, np.array([thermal_scattering_rate(T, m2) for T in Ts]))
plt.yscale("log")
plt.xscale("log")
```
